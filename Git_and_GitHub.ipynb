Qué es **GIT**

Git es una herramienta que te ayuda a guardar y organizar el trabajo que haces en archivos, permitiéndote guardar diferentes versiones de esos archivos con el tiempo. Es como un "historial de cambios" que puedes consultar, revertir o compartir con otras personas que estén trabajando contigo.

**Repositorio (Repository)**

Es el espacio donde se almacena y gestiona el historial de cambios de un proyecto. Puede ser local (en tu máquina) o remoto (en plataformas como GitHub, GitLab o Bitbucket).

**Commit**

Un "snapshot" o punto de control que registra los cambios realizados en los archivos. Es una unidad fundamental en Git que incluye un mensaje descriptivo sobre las modificaciones realizadas.

**Branch (Rama)**

Una línea de desarrollo independiente que permite trabajar en diferentes características o correcciones sin afectar el código principal. La rama principal por defecto se llama generalmente main o master.

**Merge**

Es el proceso de combinar cambios de diferentes ramas en una sola. Puede ser directo o requerir la resolución de conflictos si los cambios en las ramas son incompatibles.

**Staging Area**

Es el área de preparación donde los cambios se agregan antes de hacer un commit. Permite seleccionar qué modificaciones incluir en el próximo commit.

**Remote**

Hace referencia a las versiones remotas del repositorio, alojadas en servidores o plataformas externas. Por ejemplo, origin es el nombre comúnmente usado para el repositorio remoto principal.

**Pull**

Combina dos acciones: descargar (fetch) los cambios del repositorio remoto y fusionarlos con el repositorio local. Es útil para mantener el repositorio actualizado.

**Push**

Envía los commits locales al repositorio remoto, actualizando su historial con los cambios realizados.

**Checkout**

Permite moverse entre ramas o restaurar versiones anteriores del código. Por ejemplo, cambiar a una rama específica o volver a un commit específico.

**Tag**

Una etiqueta utilizada para marcar commits específicos, generalmente para señalar versiones importantes del proyecto (como v1.0).
